```python

class Card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank= rank
    def __str__(self):
        return self.rank +' of '+self.suit
```

```python

class Deck:
    
    def __init__(self):
    #creating the deck and cards at the same time
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))

    #shuffling the cards
    def shuffle(self):
        self.deck = list(set(self.deck))
        return self.deck
        
    #pulling a card out of deck
    def deal(self):
        pulled_card = random.choice(self.deck)
        self.deck.remove(pulled_card)
        return pulled_card
```
